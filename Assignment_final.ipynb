.csv files in this archive are in the following format:
The first row is the initial time of the session expressed as unix timestamp in UTC.
The second row is the sample rate expressed in Hz.

TEMP.csv
Data from temperature sensor expressed degrees on the Celsius (C) scale.

EDA.csv
Data from the electrodermal activity sensor expressed as microsiemens (S).

BVP.csv
Data from photoplethysmograph.

ACC.csv
Data from 3-axis accelerometer sensor. The accelerometer is configured to measure acceleration in the range [-2g, 2g]. Therefore the unit in this file is 1/64g.
Data from x, y, and z axis are respectively in first, second, and third column.

IBI.csv
Time between individuals heart beats extracted from the BVP signal.
No sample rate is needed for this file.
The first column is the time (respect to the initial time) of the detected inter-beat interval expressed in seconds (s).
The second column is the duration in seconds (s) of the detected inter-beat interval (i.e., the distance in seconds from the previous beat).

HR.csv
Average heart rate extracted from the BVP signal.The first row is the initial time of the session expressed as unix timestamp in UTC.
The second row is the sample rate expressed in Hz.


tags.csv
Event mark times.
Each row corresponds to a physical button press on the device; the same time as the status LED is first illuminated.
The time is expressed as a unix timestamp in UTC and it is synchronized with initial time of the session indicated in the related data files from the corresponding session.

```python
from scipy.signal import medfilt
import matplotlib.pyplot as plt
def peak_enhancement(bvp_signal, lower_limit=0, upper_limit=1024):
    # Apply a median filter to remove outliers
    filtered_signal = medfilt(bvp_signal)

    # Calculate the range of the filtered signal
    signal_range = np.ptp(filtered_signal)

    # Apply the peak enhancement formula
    penh_signal = upper_limit * ((filtered_signal - np.min(filtered_signal)) / signal_range) + lower_limit

    return penh_signal
bvp_signal = new_BVP_df['BVP']

bvp_signal_p_e = peak_enhancement(bvp_signal)
plt.plot(bvp_signal_p_e, label='peak enhancement')
plt.plot(bvp_signal, label ='ordinary')
plt.legend()
```

```python

def moving_average(signal, window_size):
    """
    Calculates the moving average of a signal using convolution.

    Parameters:
    signal (ndarray): Input signal.
    window_size (int): Size of the moving window.

    Returns:
    ndarray: Moving average of the input signal.
    """

    # Create the moving average window
    window = np.ones(window_size) / window_size

    # Use convolution to calculate the moving average
    ma_signal = np.convolve(signal, window, mode='same')

    return ma_signal
```

```python
import numpy as np
from scipy.signal import butter, filtfilt

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    """
    Butterworth bandpass filter to a signal with a given low cutoff frequency (fl) and high cutoff frequency (fh)

    Args:
        data (_type_): the signal that is going to be filtered
        lowcut (_type_): lower cutoff frequency
        highcut (_type_): higher cutoff frequency
        fs (_type_): sampling rate of our signal
        order (int, optional): the order of Butterworth filter. Defaults to 5.

    Returns:
        _type_: filtered signal
    """
    nyquist_freq = 0.5 * fs
    low = lowcut / nyquist_freq
    high = highcut / nyquist_freq
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y
```

```python
from scipy import signal

def butter_lowpass_filter(signal_data, cutoff_freq, order=4, fs=64):
    """Applies a low-pass filter to the given signal data.
    
    Args:
    - signal_data : array-like
        One-dimensional input data to be filtered.
    - cutoff_freq : float
        Cutoff frequency of the filter in Hz.
    - order : int (optional)
        Order of the Butterworth filter (default is 4).
    - fs : int (optional)
        Sampling rate of the input data (default is 1000).
        
    Returns:
    - filtered_signal : numpy.ndarray
        Filtered output signal with the same shape and dimension as the input data.
    """
    # Create a Butterworth low-pass filter transfer function
    b, a = signal.butter(order, cutoff_freq/(fs/2), 'low')
    
    # Apply the filter on the signal data
    filtered_signal = signal.filtfilt(b, a, signal_data)
    
    return filtered_signal
```

```python
import matplotlib.pyplot as plt

def plot_signal(dict_of_signals, t, start_time, end_time):
   


    """
        Plotting signals that are provided in dict_of_signals

        Args:
            dict_of_signals (dict): a dictionary whose keys are an string representing the name of a signal and values are the signals itself

            t (int): the length of signals. Signals should be of the same size

            start_time (int): The start point of plot

            end_time (int): The end point of plot
        
    """
       

    t_sliced = t[start_time:end_time]
    plt.figure(figsize=(10, 5))

    for _n, _s in dict_of_signals.items():
        _s_sliced = _s[start_time:end_time]
        signal_name = _n 
        plt.plot(t_sliced, _s_sliced, label=signal_name)
   
        
    plt.xlabel('Time (1/64s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import seaborn as sns
from scipy.signal import medfilt
import matplotlib.pyplot as plt

def hampel_filter(bvp_signal, window_size=17, n_sigma=3):
    # Calculate the median of the signal using a median filter
    med_signal = medfilt(bvp_signal, kernel_size=window_size)

    # Calculate the absolute deviations from the median
    deviations = np.abs(bvp_signal - med_signal)

    # Calculate the median absolute deviation (MAD)
    mad = np.median(deviations)

    # Calculate the threshold for outlier detection
    threshold = n_sigma * mad

    # Detect outliers and replace them with the median value
    outlier_mask = deviations > threshold
    outlier_indices = np.where(outlier_mask)[0]
    bvp_signal[outlier_mask] = med_signal[outlier_mask]

    return bvp_signal, outlier_indices

# Generate a sample BVP signal
bvp = new_BVP_df['BVP']
filtered_signal, outlier_indices = hampel_filter(bvp)

_slice = np.arange(64*60*18,64*60*20)
t = np.arange(len(bvp))[_slice]
outlier_indices_slice = [j for j in outlier_indices if j in _slice]
bvp_signal_sliced = bvp[_slice]
filtered_signal_sliced = filtered_signal[_slice]
# Apply the Hampel filter to remove outliers


# Create a Seaborn line plot of the original signal and filtered signal
sns.set_style("whitegrid")
sns.lineplot(x=t, y=bvp_signal_sliced, color='blue', label="BVP signal")
sns.lineplot(x=t, y=filtered_signal_sliced, color='red', label="Filtered signal")
sns.scatterplot(x=outlier_indices_slice , y=bvp_signal_sliced[outlier_indices_slice], color='purple', label='Outliers')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("BVP signal with Hampel filter applied")
plt.legend()
sns.despine()
plt.show()


# Print the indices of the outliers
print("Outlier indices:", outlier_indices)


```

```python
columns_dict = {
    'ACC': ['X', 'Y', 'Z' ],
    'BVP': ['BVP'], 
    'EDA': ['EDA',],
    'HR': ['HR'],
    'TEMP': ['TEMP'],
}


def process_df(files_path):
    '''Takes a filePath of a csv file and return the new dataFrame for that file so that it has a new column named "datetime" 
    '''
    df = pd.read_csv(files_path, header=None)
    start_timestamp = df.iloc[0,0]
    sample_rate = df.iloc[1,0]

    new_df = pd.DataFrame(df.iloc[2:].values, columns=columns_dict[files_path.stem])
    #new_df['id'] =  files_path.name[-3:]
    new_df['datetime'] = [(start_timestamp + i/sample_rate) for i in range(len(new_df))]
    return new_df

def label_generator( datetime, tags):

    '''to generate label for timestamp datetime  based on the fact that the interval number that it lies on is even or odd
    '''


    if datetime < tags[0]:
        interval = 0

    elif datetime >= tags[-1]:
        interval = 20 # Just a big number

    else:
        for i in range(len(tags)-1):
            if tags[i] <= datetime < tags[i+1]:
                interval =  i + 1

    if interval < 6:
        if (interval % 2) == 0:
            return 0
        else:
            return 1
    else:
        return 0           

df_names_list = [] # --> to get the name of all data frame that we just created
for _dir in dir_list_tmp:
    # Getting the id from path
    participant_id = int(str(_dir)[-2:])
    
    # Merging different signals
    desired_signals = ['ACC.csv', 'BVP.csv', 'EDA.csv', 'HR.csv', 'TEMP.csv']
    files_path_list = [file_path  for file_path in _dir.iterdir() if file_path.name in desired_signals]
    
    for f in files_path_list:
        globals()[f'new_{f.stem}_df'] = process_df(f)

    merged_df = new_ACC_df.merge(new_EDA_df, on='datetime', how='outer' )
    merged_df = merged_df.merge(new_BVP_df, on='datetime', how='outer')
    merged_df = merged_df.merge(new_HR_df, on='datetime', how='outer')
    merged_df = merged_df.merge(new_TEMP_df, on='datetime', how='outer')
    
    # Dropping NaN values
    merged_df_drop = merged_df.dropna(axis=0)


    # Adding label
    tags_path = [t for t in _dir.iterdir() if t.name[0] == 't'] # --> only one file name in each folder that starts with 't'
    tags_df = pd.read_csv(tags_path[0], header=None)
    
    tags =  np.sort(tags_df.values)
    merged_df_drop['Label'] = merged_df_drop['datetime'].apply(lambda x: label_generator(x, tags))


    
    # Adding 'id' column
    merged_df_drop['id'] = participant_id

    # Specifying a specific name for each merged dataFrame
    globals()[f'df_{participant_id}'] = merged_df_drop

    df_names_list.append(f'df_{participant_id}')






    
```

```python
# Calculating position, velocity and accelerate for each dataFrame before concatenating all of them
# Adding position column
def position(x, y, z):
    return np.sqrt(x**2 + y**2 + z**2)
    
# Adding velocity column
def velocity(x, y, z):
    return np.sqrt(x**2 + y**2 + z**2)  

# Adding accelerate column
def accelerate(x, y, z):
    return np.sqrt(x**2 + y**2 + z**2)  

for df_name in df_names_list:
    # Adding position column

    # Creating velocity components
    globals()[df_name]['V_X'] = globals()[df_name]['X'].diff().shift(-1)
    globals()[df_name]['V_Y'] = globals()[df_name]['Y'].diff().shift(-1)
    globals()[df_name]['V_Z'] = globals()[df_name]['Z'].diff().shift(-1)

    # Creating accelerate components
    globals()[df_name]['A_X'] = globals()[df_name]['X'].diff().shift(-2)
    globals()[df_name]['A_Y'] = globals()[df_name]['Y'].diff().shift(-2)
    globals()[df_name]['A_Z'] = globals()[df_name]['Z'].diff().shift(-2)



    # apply the position, velocity and accelerate function 
    globals()[df_name]['position'] = globals()[df_name].apply(lambda row: position(row['X'], row['Y'], row['Z']), axis=1)
    globals()[df_name]['velocity'] = globals()[df_name].apply(lambda row: velocity(row['V_X'], row['V_Y'], row['V_Z']), axis=1)
    globals()[df_name]['accelerate'] = globals()[df_name].apply(lambda row: accelerate(row['A_X'], row['A_Y'], row['A_Z']), axis=1)


    
```

```python

# concatenate dataFrames vertically using concat() function with axis=0
df_raw_complete = pd.concat([globals()[df_name] for df_name in df_names_list], axis=0)
df_raw_complete
```
